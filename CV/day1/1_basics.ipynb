```python
import numpy as np
```

```python
imgz = np.zeros([500,500], dtype=np.uint8)
plt.imshow(imgz, cmap='gray');
```

```python
# changing image's mid row
imgz[250,:] = 123
plt.imshow(imgz, cmap='gray');
```

```python
# changing image's mid col
imgz[: ,250] = 123
plt.imshow(imgz, cmap='gray');
```

```python
# changing image's block
imgz[ : 100, : 100] = 123
imgz[200: 301, : 100] = 123
imgz[400: , : 100] = 123
imgz[ : 100, 200:300] = 123
imgz[:100 ,400:] = 123
imgz[200: 301,200:300] = 123
imgz[400 : ,200:300 ] = 123
imgz[200:300, 400:] = 123
imgz[400:, 400:] = 123
plt.imshow(imgz, cmap='gray');
```

```python
img_sam = np.zeros([10000], dtype=np.uint8)
```

```python
for i in range(10000):
    img_sam[i] = np.random.randint(0, 255)
```

```python
img_sam = img_sam.reshape(100, 100)
plt.imshow(img_sam, cmap='gray')
```

```python
img_chess = np.zeros([8, 8])

for i in range(img_chess.shape[0]):
    for j in range(img_chess.shape[1]):
        if (i + j)%2 == 0:
            img_chess[i][j] = 1

plt.title('Chess Board')
plt.imshow(img_chess, cmap='gray');
```

```python
img_gray = cv2.imread(dogesh,0)
shape = img_gray.shape

for i in range(shape[0]):
    for j in range(shape[1]):
        if i == j or (i + j) == shape[0]:
            img_gray[i, j] = 0


plt.imshow(img_gray, cmap='gray');
```
