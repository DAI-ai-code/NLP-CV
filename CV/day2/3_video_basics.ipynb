extract all images and save in a folder

```python
vid = cv2.VideoCapture('../input/images/nature.mp4')
i = 1
while True:
    ret, frame = vid.read()
    if not ret:
        break
    cv2.imwrite(f'../input/frames/frame{i}.jpg', frame)  # can also use "./frames/frame%d.jpg" with %i
    i += 1
```

```python
 # trigger frame (at a particular time)

def frame_finder(timestep):
    fps = int(vid.get(cv2.CAP_PROP_FPS))


    hrs, mins, secs = timestep.split(':')
    total_secs = int(hrs) * 3600 + int(mins) * 60 + float(secs) 
    trigger_frame = int(total_secs * fps)


    i = 1
    while True:
        ret, frame = vid.read()
        if not ret:
            break

        if i == trigger_frame:
            cv2.imwrite('./frames/trigger_frame_%s.jpg' %timestep, frame)
        
        i += 1

```

```python
time = "00:00:45.00" 
frame_finder(time)
```

```python
hrs, mins, secs = time.split(':')
trigger_frame = (int(hrs) * 3600 + int(mins) * 60 + float(secs)) * fps
print(trigger_frame)
vid.set(1, trigger_frame)  # 1 here means to jump to a particular frame with number trigger_frame
```

```python
ret, frame = vid.read()
cv2.imshow("Trigger frame", frame)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
