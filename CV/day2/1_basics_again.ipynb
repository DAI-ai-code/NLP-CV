```python
import matplotlib.pyplot as plt
```

```python

dogesh = '../input/images/dog.jpg'
img_plt = plt.imread(dogesh)
plt.imshow(img_plt);
```

```python
img_new = img_plt.copy()

for i in range(img_plt.shape[0]):
    for j in range(img_plt.shape[1]):
        if i==j or (i+j)==img_plt.shape[1]:
            img_new[i,j] = 0

plt.imshow(img_new)
```

```python
import numpy as np

orange = (255, 165, 0)
white = (255,255,255)
green = (0,120,0)

flag = np.zeros([3,4,3], dtype=int)

flag[0] = orange
flag[1] = white
flag[2] = green

plt.imshow(flag)
```

#### converting color image to grayscale image

```python
img_plt.shape
```

```python
import cv2

img_cv = cv2.imread('../input/images/dog.jpg')

plt.imshow(img_cv)
```

```python
img_plt.shape
```

```python
img_gray = cv2.cvtColor(img_plt, cv2.COLOR_RGB2GRAY)
plt.imshow(img_gray, cmap='gray');   # cmap works only for grayscale images
```

```python
img_plt
```

```python
r = img_plt[:, :, 0]  # all rows, all columns, channel 0 for red
g = img_plt[:, :, 1]  # ---------------------- channel 1 for green
b = img_plt[:, :, 2]  # ---------------------- channel 2 for blue
r
```

```python
plt.figure(figsize=(16, 9))
plt.subplot(131); plt.imshow(r, cmap='gray')
plt.subplot(132); plt.imshow(g, cmap='gray')
plt.subplot(133); plt.imshow(b, cmap='gray')
```

```python
plt.figure(figsize=(16, 9))
plt.subplot(131); plt.imshow(r, cmap='Reds')
plt.subplot(132); plt.imshow(g, cmap='Greens')
plt.subplot(133); plt.imshow(b, cmap='Blues')
```

```python
img_g = img_plt.copy()

avg = (r+b+g)//3

print(avg.shape,'\n')

print(avg)

plt.imshow(avg, cmap='gray')
```

#### Splitting and merging

```python
b, g, r = cv2.split(img_cv)
b.shape, g.shape, r.shape
```

```python
img_rgb = cv2.merge([r, g, b])
img_gbr = cv2.merge([g, b, r])
```

```python
plt.subplot(131)
plt.imshow(img_plt)
plt.subplot(132)
plt.imshow(img_rgb)
plt.subplot(133)
plt.imshow(img_gbr)
```
