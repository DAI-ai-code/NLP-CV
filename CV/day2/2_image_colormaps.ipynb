```python
import cv2
import matplotlib.pyplot as plt
```

```python
dogesh = '../input/images/dog.jpg'
```

0 mean lightest hai, 255 means darkest hai

```python
img = cv2.imread(dogesh,0)


plt.subplot(141)
plt.imshow(img, cmap=plt.colormaps['PuBu'].reversed())
plt.subplot(142)
plt.imshow(img, cmap='PuBu')
plt.subplot(143)
plt.imshow(img, cmap='BuPu')
plt.subplot(144)
plt.imshow(img, cmap=plt.colormaps['BuPu'].reversed())
```

```python
img = plt.imread(dogesh)
img_norm = img/255
plt.imshow(img_norm)
```
