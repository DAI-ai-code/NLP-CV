```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = plt.imread("../input/images/cat_damaged.png")
plt.imshow(image)
```

```python
height, width = image.shape[:-1]
```

```python
mask = np.zeros([height, width, 3], dtype=np.uint8)
```

```python
for i in range(height):
    for j in range(width):
        if image[i,j].sum() == 0:
            mask[i,j] =255

```

```python
mask = cv2.cvtColor(mask, cv2.COLOR_BGR2GRAY)
```

```python
#inpaint
image = cv2.imread("../input/images/cat_damaged.png")
dst = cv2.inpaint(image, mask, 3, cv2.INPAINT_NS)
```

```python
cv2.imshow("resurrected", dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
