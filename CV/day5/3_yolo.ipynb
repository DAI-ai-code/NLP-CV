```python
import torch
import matplotlib.pyplot as plt
import cv2
from PIL import Image
import numpy as np
```

```python
# function to load image
def load_image(image):
  image = Image.open(image)
  return image
```

```python
# load yolo model
model = torch.hub.load("ultralytics/yolov5", "yolov5s")
```

```python
# define predict function
def predict(image):
  # load image
  image = load_image(image)

  # run model inference
  results = model(image)

  # show detected objects
  print('Detected objects: ', results.names)

  results.show()
```

```python
predict('../input/images/jungle.jpg')
```

```python
# read all video
vid = cv2.VideoCapture('../input/images/robots.mp4')

while vid.isOpened():
    ret, frame = vid.read()
    if not ret:
        print('End of video stream')
        break
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = model(frame_rgb)
    results.render()
    # results.ims?
    final_frame = cv2.cvtColor(results.ims[0], cv2.COLOR_BGR2RGB)
    cv2.imshow('YOLOv5 Object Detection', final_frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

vid.release()
cv2.destroyAllWindows()
```
