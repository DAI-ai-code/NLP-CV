```python
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input, decode_predictions
from tensorflow.keras.preprocessing.image import load_img, img_to_array
```

```python
img = load_img('hitu.jpg')
```

```python
type(img)
```

```python
imgg = img_to_array(img)
imgg.shape
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
import cv2
imgg = cv2.resize(imgg, (224,224))

imgg.shape
```

```python
imgg = preprocess_input(imgg)
```

```python
help(VGG16)
```

```python
model = VGG16()
```

```python
result = model.predict(imgg.reshape(1,224,224,3))
```

```python
decode_predictions(result)
```
