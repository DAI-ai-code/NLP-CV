# Binary Thresholding
(thresholding works only for grayscale)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("../input/images/dog.jpg",0)
img_new = img.copy()

threshold = 130
for i in range(img_new.shape[0]):
    for j in range(img_new.shape[1]):
        if img_new[i][j] > threshold:
            img_new[i][j] = 255
        else:
            img_new[i][j] = 0

plt.figure(figsize=(16,9))
plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.subplot(122)
plt.imshow(img_new, cmap='gray')
```

```python
## OR
img_newer = img > 130
plt.imshow(img_newer, cmap='gray')
```
