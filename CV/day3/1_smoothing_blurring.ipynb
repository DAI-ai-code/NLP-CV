# Averaging kernel

```python
img = cv2.imread('../input/images/dog.jpg')
kernel = np.zeros([9,9]) + 1/81
```

```python
blurred_img = cv2.filter2D(img, ddepth=-1, kernel=kernel)
blurred_img = np.hstack([img,blurred_img])
```

```python
cv2.imshow("BI",blurred_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Gaussian Kernel

```python
kernel_size = 7
kernel = cv2.getGaussianKernel(kernel_size, sigma=1.23)
```

```python
gaussian_blurred_img = cv2.filter2D(img, ddepth=-1, kernel=kernel)
gaussian_stack = np.hstack([img, gaussian_blurred_img])
```

```python
cv2.imshow("og vs gaussian blurred", gaussian_stack)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Median blur

```python
img_gray = cv2.imread("../input/images/dog.jpg",0)
median_img = cv2.medianBlur(img_gray, 15) # 15 is size of kernel
```

```python
plt.figure(figsize=(16,9))
plt.subplot(121)
plt.imshow(img_gray, cmap='gray')
plt.subplot(122)
plt.imshow(median_img, cmap='gray')
```

### blurring small patch

```python
start_row, end_row, start_col, end_col = 100, 250, 100, 250

patch = img_gray[start_row: end_row, start_col: end_col]
plt.imshow(patch, cmap='gray')
```

```python
# edited img
img_edit = img_gray.copy()
img_edit[start_row: end_row, start_col: end_col] = 0

plt.imshow(img_edit, cmap='gray')
```

```python
blurred_patch = cv2.medianBlur(patch, 15)

plt.imshow(blurred_patch, cmap='gray')
```

```python
img_edit[start_row: end_row, start_col: end_col] = blurred_patch

plt.imshow(img_edit, cmap='gray')
```

```python
plt.figure(figsize=(16,9))
plt.subplot(121)
plt.imshow(img_gray, cmap='gray')
plt.subplot(122)
plt.imshow(img_edit, cmap='gray')
```
