```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

# Harris Corner Detection 

```python
img = cv2.imread('../input/images/circle.png',0)

corners = cv2.cornerHarris(img, blockSize=2, ksize=3, k=0.02)



plt.imshow(corners, cmap='gray')
# plt.figure(figsize=(12, 10))
# plt.subplot(121)
# plt.title('original')
# plt.imshow(img, cmap='gray')

# plt.subplot(122)
# plt.title('sobel combined')
# plt.imshow(img_canny, cmap='gray')
```

```python
corners.min(), corners.max(), np.median(corners)
```

```python
img_new = img.copy()
img_new[corners>0.01 * corners.max()] = 255

plt.imshow(img_new, cmap='gray')
```

```python
corn = corners.copy()

corn[corn < 0.001] = np.median(corn)
corn[corn > 0] = 255
plt.imshow(corn, cmap='gray')
```
