```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../input/images/lenna.png', 0)
```

```python
sobelx = cv2.Sobel(img, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=3)
sobely = cv2.Sobel(img, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=3)
```

```python
sobelx.shape, sobelx.dtype
```

```python
# add both x and y 
final_sobel = cv2.add(sobelx, sobely) # this highlights edges
```

```python
# now apply thresholding
thresh, binary_sobel = cv2.threshold(final_sobel, 100, 255, cv2.THRESH_BINARY)
```

```python
thresh
```

```python
plt.subplot(121)
plt.title('original')
plt.imshow(img, cmap='gray')

plt.subplot(122)
plt.title('sobel')
plt.imshow(binary_sobel, cmap='gray')
```

```python
thresh, sobelx_img = cv2.threshold(sobelx, 60, 255, cv2.THRESH_BINARY)
thresh, sobely_img = cv2.threshold(sobely, 60, 255, cv2.THRESH_BINARY)
```

```python
plt.subplot(141)
plt.title('original')
plt.imshow(img, cmap='gray')

plt.subplot(142)
plt.title('sobel combined')
plt.imshow(binary_sobel, cmap='gray')


plt.subplot(143)
plt.title('sobel x')
plt.imshow(sobelx_img, cmap='gray')


plt.subplot(144)
plt.title('sobel y')
plt.imshow(sobely_img, cmap='gray')
```

```python
img = cv2.imread('../input/images/circle.png', 0)

sobelx = cv2.Sobel(img, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=3)
sobely = cv2.Sobel(img, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=3)
# add both x and y 
final_sobel = cv2.add(sobelx, sobely) # this highlights edges
# now apply thresholding
thresh, binary_sobel = cv2.threshold(final_sobel, 100, 255, cv2.THRESH_BINARY)


thresh, sobelx_img = cv2.threshold(sobelx, 100, 255, cv2.THRESH_BINARY)
thresh, sobely_img = cv2.threshold(sobely, 100, 255, cv2.THRESH_BINARY)


plt.figure(figsize=(12, 10))
plt.subplot(141)
plt.title('original')
plt.imshow(img, cmap='gray')

plt.subplot(142)
plt.title('sobel combined')
plt.imshow(binary_sobel, cmap='gray')


plt.subplot(143)
plt.title('sobel x')
plt.imshow(sobelx_img, cmap='gray')


plt.subplot(144)
plt.title('sobel y')
plt.imshow(sobely_img, cmap='gray')
```
