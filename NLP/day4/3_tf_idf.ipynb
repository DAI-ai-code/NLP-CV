```python
sent1 = 'It is good practice for us.'
sent2 = 'It was also good to know about it.'
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
```

```python
new_text = tfidf.fit_transform([sent1, sent2])
```

```python
new_text.toarray()
```

```python
import pandas as pd

df = pd.DataFrame(new_text.toarray(), columns=tfidf.get_feature_names_out())
df
```

```python
tfidf = TfidfVectorizer(stop_words='english')
```

```python
new_text = tfidf.fit_transform([sent1, sent2])
```

```python
df = pd.DataFrame(new_text.toarray(), columns=tfidf.get_feature_names_out())
```

```python
df
```

Reading sample sentences

```python
file_name = '../input/sample_sentences.txt'

with open(file=file_name, mode='r') as file:
    data = file.readlines()

data
```

```python
tfidf = TfidfVectorizer()
new_text = tfidf.fit_transform(data)
new_text
```

```python
df = pd.DataFrame(new_text.toarray(), columns=tfidf.get_feature_names_out())
df
```

```python
tfidf = TfidfVectorizer(stop_words='english')
new_text = tfidf.fit_transform(data)
```

```python
df = pd.DataFrame(new_text.toarray(), columns=tfidf.get_feature_names_out())
df
```
