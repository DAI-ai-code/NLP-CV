```python
import pyttsx3
```

```python
engine = pyttsx3.init()

# set the properties

engine.setProperty('rate', 120)
engine.setProperty('volume', 1)
```

```python
text = 'Arnav and Aman your keyboard will be swapped.'
engine.say(text)
engine.runAndWait()
# restart kernel
```
