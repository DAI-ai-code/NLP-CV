
 2. Remove punctuation symbols

```python
clean = [word for word,pos in tagged if pos.isalnum()]
clean[:10]
```


 3. Delete the stopwords

```python
from nltk.corpus import stopwords
swords = stopwords.words('english')
cleaned = [i for i in pos_tag(word_tokenize(text)) if i not in swords]
cleaned
```


 4. Perform morphological analysis on all the nouns

```python
nouns = [i for i,pos in pos_tag(word_tokenize(text)) if pos.startswith('N')]
nouns

from nltk.stem import WordNetLemmatizer
lem = WordNetLemmatizer()

morphs = [lem.lemmatize(w) for w in nouns]
morphs
```


 5. Find top three words in the data

```python
from collections import Counter
from nltk.corpus import stopwords
no_stop = stopwords.words('english')
clean = [ i for i in clean if i not in no_stop]
freq = Counter(clean)
freq.most_common(3)
```
