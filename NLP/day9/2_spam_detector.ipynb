```python
from tkinter import *
import joblib
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
top = Tk()
top.title("SPAM Message Detector")
top.geometry("800x500")

ps = PorterStemmer()
swords = stopwords.words('english')

# cleaning function
def clean_text(sent: str):
    # tokenize 
    tokens1 = word_tokenize(sent)
    tokens2 = [token for token in tokens1 if token.isalpha()]
    tokens3 = [ps.stem(token.lower()) for token in tokens2 if token.lower() not in swords]
    return tokens3


def check():
    message = t.get("1.0", END)
    mes_new = preprocessor.transform([message])
    pred = classifier.predict(mes_new)
    if pred[0] == 'ham':
        result.config(text="Message is NOT SPAM")
    else:
        result.config(text="Message is SPAM")

preprocessor = joblib.load('../day4/preprocessor.model')
classifier = joblib.load('../day4/classifier.model')

Label(top, text="Enter your message here.").place(x=100, y=10)

t = Text(top)
t.place(x=20, y=40)

Button(top, text="Check Message", command=check).place(x=100, y=450)

result = Label(top)
result.place(x=100, y=470)

top.mainloop()
```
