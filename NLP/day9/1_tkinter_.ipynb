```python
from tkinter import *
```

```python
top = Tk()
top.title("My first application")
top.geometry("500x300")
top.mainloop()
```

```python
top = Tk()
top.geometry("500x300")

def find():
    num = int(e.get())
    result.config(text="Square is: " + str(num*num))  
    

l = Label(top,text="Enter number: ")
e = Entry(top)
b = Button(top,text="Find square", command=find)
result = Label(top)

l.pack()
e.pack()
b.pack()
result.pack()
top.mainloop()
```

# PLACE

```python
top = Tk()
top.geometry("500x300")

def find():
    num = int(e.get())
    result.config(text="Square is: " + str(num*num))  
    

l = Label(top,text="Enter number: ")
e = Entry(top)
b = Button(top,text="Find square", command=find)
result = Label(top)

l.place(x=50,y=50)
e.place(x=50,y=80)
b.place(x=50,y=110)
result.place(x=50,y=170)
top.mainloop()
```

### Calculator

```python
top = Tk()
top.geometry('500x300')

def add():
    n1 = int(num_1_entry.get())
    n2 = int(num_2_entry.get())
    res =  n1 + n2
    result.config(text='Result is : ' + str(res))    

def subtract():
    n1 = int(num_1_entry.get())
    n2 = int(num_2_entry.get())
    res =  n1 - n2
    result.config(text='Result is : ' + str(res))


def multiply():
    n1 = int(num_1_entry.get())
    n2 = int(num_2_entry.get())
    res =  n1 * n2
    result.config(text='Result is : ' + str(res))    


def divide():
    n1 = int(num_1_entry.get())
    n2 = int(num_2_entry.get())
    res = n1 / n2
    result.config(text='Result is : ' + str(res))




Label(top, text='Number 1 ').place(x=50, y=50)
num_1_entry = Entry(top)

Label(top, text='Number 2 ').place(x=50, y = 100)
num_2_entry = Entry(top)

num_1_entry.place(x=50, y=80)
num_2_entry.place(x=50, y=130)

Button(top, text='+', command=add).place(x=50, y= 160)
Button(top, text='-', command=subtract).place(x=90, y= 160)
Button(top, text='*', command=multiply).place(x=130, y= 160)
Button(top, text='/', command=divide).place(x=170, y= 160)

result = Label(top, text='Result is : ')
result.place(x=50, y = 200)

top.mainloop()
```
