```python
import string
import nltk

from nltk.tokenize import wordpunct_tokenize

sent = 'Hello friends! How are you doing? Welcome to "Python" Programming.'
ans = ''
for i in wordpunct_tokenize(sent):
    if i not in string.punctuation:
        ans += i  + " "
ans
```

```python
ans = [i for i in wordpunct_tokenize(sent) if i not in string.punctuation]
" ".join(ans)
```

#### Stopwords removal

```python
from nltk.corpus import stopwords

swords = stopwords.words('english')
removed = [i for i in ans if i.lower() not in swords]
" ".join(removed)
```

#### Stemming

```python
words1 = 'consultant', 'consult', 'consultants', 'consulting', 'consultancy', 'consulted', 'consultative'
words2 = 'management', 'managed', 'manage'
words3 = 'big', 'bigger', 'biggest'
words4 = 'association', 'associating', 'associated', 'associations'
```

```python
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
ps.stem(words2[0])
```

```python
for word in words1:
    print(ps.stem(word))
```

```python
for word in words2:
    print(ps.stem(word))
```

```python
for word in words3:
    print(ps.stem(word))
```

```python
for word in words4:
    print(ps.stem(word))
```

```python
for word in ['unmanageable', 'unbreakable']:
    print(ps.stem(word))
```

```python
for word in ['go', 'gone', 'went']:
    print(ps.stem(word))
```

#### Lemmatization

```python
from nltk.stem import WordNetLemmatizer

wnl = WordNetLemmatizer()
```

```python
words5 = 'go', 'gone', 'went', 'going'
```

```python
for word in words5:
    print(wnl.lemmatize(word, pos='v'))
```

```python
for word in words1:
    print(wnl.lemmatize(word, pos='v'))
```

```python
for word in words1:
    print(wnl.lemmatize(word, pos='n'))
```

```python
for word in words2:
    print(wnl.lemmatize(word, pos='v'))
```

```python
for word in words2:
    print(wnl.lemmatize(word, pos='n'))
```

```python
for word in words4:
    print(wnl.lemmatize(word, pos='v'))
```

```python
for word in words4:
    print(wnl.lemmatize(word, pos='n'))
```

```python
print(wnl.lemmatize('is', pos='v'))
print(wnl.lemmatize('children', pos='n'))
print(wnl.lemmatize('worst', pos='a'))
```
