```python
from nltk.tokenize import word_tokenize
from nltk import pos_tag, RegexpParser
```

```python
# create chunk grammar for regex
grammar = "NP:{<DT><JJ><NN>}"
```

```python
sent = 'It was a beautiful game by him. An amazing play was witnessed.'
```

```python
chunker = RegexpParser(grammar)
```

```python
tags = pos_tag(word_tokenize(sent))
```

```python
tags
```

```python
tree = chunker.parse(tags)
```

```python
tree.draw()
```

```python
tree
```

Verb Phrase Combination

```python
sent1 = 'They were watching the game. It was just thrilling.'
```

```python
grammar1 = "VP:{<PRP><VBD>}"
```

```python
parser = RegexpParser(grammar1)
```

```python
tags = pos_tag(word_tokenize(sent1))
```

```python
tree1 = parser.parse(tags)
```
