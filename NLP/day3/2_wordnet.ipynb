### Q. Extract specific POS from the word

```python
verbs = set()
nouns = set()
adjts = set()

for w in wordnet.synsets('sound'):
    if w.pos() == 'n':
        nouns.add(w.lemmas()[0].name())
    if w.pos() == 'v':
        verbs.add(w.lemmas()[0].name())
    if w.pos() == 'a':
        adjts.add(w.lemmas()[0].name())

print(verbs)
print(adjts)
print(nouns)

 
```

### Extract antonyms

```python
antonyms = set()

for word in wordnet.synsets('peak'):
    for lemma in word.lemmas():
        if lemma.antonyms():
            antonyms.add(lemma.antonyms()[0].name())

antonyms
```

### Extract hypernyms

```python
word = 'isis'
synsets = wordnet.synsets(word)

for x in synsets:
    hypernyms = x.hypernyms()
    for h in hypernyms:
        print(word, ' -- hypernym of -- ', h.name())
```

### Extract meronymns

```python
word = 'car'
synsets = wordnet.synsets(word)

for x in synsets:
    meronyms = x.part_meronyms()
    for m in meronyms:
        print(word, 'meronyms of', m.name())
```

```python
word = 'car'
synsets = wordnet.synsets(word)

for x in synsets:
    holo = x.part_holonyms()
    for m in holo:
        print(word, 'holo of', m.name())
```
